```python
import web_scraping
from web_scraping.scrape import extract_links, extract_recipe_details
```

```python
recipe_links = []
all_recipes = {}
base_url = "https://www.bbc.co.uk/food/recipes/a-z/{alphabet}/{page}#featured-content"
```

```python
alphabets = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'y', 'z', '0-9']
```

```python
for alphabet in alphabets:
    recipe_links.extend(extract_links(alphabet, base_url))
```

```python
with open("recipe_links.txt", "w") as f:
    for link in recipe_links:
        f.write(link + "\n")
```

```python
#with open("recipe_links.txt", "r") as f:
#    recipe_links = [line.strip() for line in f]
```

```python
total_links = len(recipe_links)

for index, url in enumerate(recipe_links, start=1):
    print(f"Processing {index}/{total_links}: {url}")  # Show progress
    
    recipe_data = extract_recipe_details(url)

    if recipe_data: 
        all_recipes.update(recipe_data)
```

```python
with open("bbc_recipes_ingredients.txt", "w") as file:
    for title, details in all_recipes.items():
        file.write(f"{title}:\n")
        file.write(f"Link: {details['link']}\n")  # Add the link
        
        for ingredient in details["ingredients"]:
            file.write(f"- {ingredient}\n")
        
        file.write("\n")  # Add a blank line between recipes
```
