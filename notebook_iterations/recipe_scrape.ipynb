```python
import requests
from bs4 import BeautifulSoup
```

```python
from datetime import datetime
print(datetime.now().strftime("%I:%M:%S %p"))
print(f"Now is:{datetime.now().strftime("%I:%M:%S %p")}")
```

```python
base_url = "https://www.bbc.co.uk/food/recipes/a-z/a/{page}#featured-content"
recipe_links = []

# Loop through pages 1 to 13
for page in range(1, 14):
    url = base_url.format(page=page)
    response = requests.get(url)

    soup = BeautifulSoup(response.text, "html.parser")

    # Extract recipe links
    for div in soup.find_all("div", class_="gel-layout__item"):
        a_tag = div.find("a", class_="promo")
        if a_tag and "href" in a_tag.attrs:
            link = "https://www.bbc.co.uk" + a_tag["href"]
            recipe_links.append(link)
```

```python
all_recipes = {}

# Find ingredients in recipe links
for url in recipe_links:
    soup = BeautifulSoup(requests.get(url).text, "html.parser")
    title = soup.select_one("h1")
    ingredients = {a.text.strip() for a in soup.select('a[data-testid="ingredient-derived-link"]')}
    
    if ingredients:
        all_recipes[title.text.strip()] = sorted(ingredients)
        print(f"✅ Scraped: {title.text.strip()} ({len(ingredients)} ingredients)")
```

```python
# Safe ingredients in a text file
with open("bbc_recipes_ingredients.txt", "w") as file:
    for recipe, ingredients in all_recipes.items():
        file.write(f"{recipe}:\n")
        for ingredient in ingredients:
            file.write(f"- {ingredient}\n")
        file.write("\n")
```
